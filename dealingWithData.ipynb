```
from sklearn.datasets import load_iris
```

```
import numpy as np
import keras
np.random.seed(10)
```

```
iris = load_iris()
print(iris.keys())
```

```
x = iris['data']
y = iris['target']
name = iris['target_names']
feature = iris['feature_names']
```

```
isample = np.random.randint(len(y),size =(5))
```

```
print(x.shape,y.shape)
```

```
print(x[isample],y[isample])
```

```
Ny = len(np.unique(y))
#one hot encoding
y = keras.utils.to_categorical(y[:],num_classes=Ny)
```

```
print("X : ",x[isample,:])
print("Y : ",y[isample])
```

```
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
X_train,X_test,Y_train,Y_test = train_test_split(x,y,test_size =0.2,random_state =1)
print(X_train.shape)
print(X_test.shape)
```

```
from sklearn.preprocessing import StandardScaler
scalar = StandardScaler()
scalar.fit(X_train) #computing mean and variance of the given data
X_train = scalar.transform(X_train)
X_test = scalar.transform(X_test)
print(X_train)
```

```
#training data
addicol = lambda x: np.concatenate((x, np.ones((x.shape[0],1))),axis=1)
Ns,Nx = X_train.shape
XX = addicol(X_train)
YY = Y_train
print(YY.shape)
w = np.linalg.inv(XX.T.dot(XX)).dot(XX.T.dot(YY))
print(w)
def evaluate(X,W,Yd):
  x = addicol(X)
  yd = np.argmax(Yd,axis =1)
  y = np.argmax(x.dot(W),axis =1)
  print("CM:")
  print(confusion_matrix(yd,y))
evaluate(X_train,w,Y_train)
evaluate(X_test,w,Y_test)
```
